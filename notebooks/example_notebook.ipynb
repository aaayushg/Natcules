```python
from descriptor_generator import calculate_descriptors
from fragmentation import fragment_molecule

# Step 1: Generate descriptors
df["descriptors"] = df["SMILES"].apply(calculate_descriptors)

# Step 2: Define a function to safely apply fragmentation and handle errors

def get_fragments(smiles, method='brics'):
    try:
        return fragment_molecule(smiles, method)
    except Exception:
        return []  # or None if you prefer

# Step 3: Apply to the DataFrame and create a new column
df["fragments"] = df["SMILES"].apply(lambda x: get_fragments(x, method="brics"))
```

```python
from enumeration import enumerate_molecules_from_fragments

# Step 4: Enumerate new molecules from the fragments
df["enumerated_molecules"] = df["fragments"].apply(
    lambda frag_list: enumerate_molecules_from_fragments(frag_list, max_output=20)
)
```

```python
from property_predictor import predict_properties

# Step 5: Predict properties for the enumerated molecules

df["predicted_properties"] = df["smiles"].apply(predict_properties)
```

```python
from rdkit import Chem
from rdkit.Chem import Draw

# List of SMILES strings for 10 plant-based molecules
smiles_list = [
    "COc1cc(C=CC(=O)CC(=O)C=Cc2ccc(O)c(OC)c2)ccc1O",  # Curcumin
    "C1=CC(=C(C=C1C2=C(C(=O)C3=C(C=C(C=C3O2)O)O)O)O)O",  # Quercetin
    "CC(C)=CCCOC(C)C",  # Linalool
    "CC(C)C1=CC=C(C=C1)O",  # Thymol
    "COC1=CC2=C(C=C1OC)N=CC3=C2C=CC4=CC=CC=C43",  # Berberine
    "CC(C)[C@@H]1CC[C@H](C)C(C)C1",  # Menthol
    "COC1=CC=C(C=C1)CC=C",  # Eugenol
    "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",  # Caffeine
    "C1=CC(=CC=C1C=CC2=CC(=CC=C2)O)O",  # Resveratrol
    "C1=CC=C(C=C1C(=O)O)O"  # Salicylic acid
]

# Corresponding names
names = [
    "Curcumin", "Quercetin", "Linalool", "Thymol", "Berberine",
    "Menthol", "Eugenol", "Caffeine", "Resveratrol", "Salicylic Acid"
]

# Convert to RDKit molecule objects and compute 2D coordinates
molecules = []
for smi in smiles_list:
    mol = Chem.MolFromSmiles(smi)
    if mol:
        Chem.rdDepictor.Compute2DCoords(mol)
        molecules.append(mol)

# Draw and save the image
img = Draw.MolsToGridImage(
    molecules,
    molsPerRow=5,
    subImgSize=(300, 300),
    #legends=names
)

# Save the image
#img.save("plant_based_molecules_10.png")
print("Image saved as 'plant_based_molecules_10.png'")
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.BRICS import BRICSDecompose

#from PIL import Image  # âœ… Make sure this is imported


# List of SMILES strings for 10 plant-based molecules
smiles_list = [
    "COc1cc(C=CC(=O)CC(=O)C=Cc2ccc(O)c(OC)c2)ccc1O",  # Curcumin
    "C1=CC(=C(C=C1C2=C(C(=O)C3=C(C=C(C=C3O2)O)O)O)O)O",  # Quercetin
    "CC(C)=CCCOC(C)C",  # Linalool
    "CC(C)C1=CC=C(C=C1)O",  # Thymol
    "COC1=CC2=C(C=C1OC)N=CC3=C2C=CC4=CC=CC=C43",  # Berberine
    "CC(C)[C@@H]1CC[C@H](C)C(C)C1",  # Menthol
    "COC1=CC=C(C=C1)CC=C",  # Eugenol
    "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",  # Caffeine
    "C1=CC(=CC=C1C=CC2=CC(=CC=C2)O)O",  # Resveratrol
    "C1=CC=C(C=C1C(=O)O)O"  # Salicylic acid
]

# Fragment molecules using BRICS and collect unique fragments
all_fragments = set()
for smi in smiles_list:
    mol = Chem.MolFromSmiles(smi)
    if mol:
        frags = BRICSDecompose(mol)
        all_fragments.update(frags)

# Convert fragment SMILES to Mol objects
fragment_mols = [Chem.MolFromSmiles(frag) for frag in all_fragments if Chem.MolFromSmiles(frag) is not None]

# Generate 2D coordinates for each fragment
for mol in fragment_mols:
    Chem.rdDepictor.Compute2DCoords(mol)

# Draw fragments
img = Draw.MolsToGridImage(
    fragment_mols,
    molsPerRow=5,
    subImgSize=(250, 250),
    #legends=[f"Frag {i+1}" for i in range(len(fragment_mols))]
)
```
